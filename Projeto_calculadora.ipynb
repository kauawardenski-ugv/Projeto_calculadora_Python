```python
# 5 "caixas" vazias para guardar histórico
# Colocamos a palavra "vazio"

mem_1 = "vazio"
mem_2 = "vazio"
mem_3 = "vazio"
mem_4 = "vazio"
mem_5 = "vazio"

# Ela serve para o programa lembrar de quem é a vez de receber o próximo resultado.
# Como ela começa valendo 1, a primeira conta feita vai direto para a gaveta 'mem_1'.
lugar_da_memoria = 1

# A chave geral do programa. Enquanto estiver como "sim", o programa continua ligado.
programa_rodando = "sim"

print("--- CALCULADORA ---")


#O MOTOR DO PROGRAMA

# O comando 'while' (enquanto) faz a nossa roda gigante girar.
# Ele repete tudo que está aqui dentro infinitamente, até a chave virar "nao" com a opcao 7 que é de sair.
while programa_rodando == "sim":
    print(" ")
    print("1 - Somar")
    print("2 - Subtrair")
    print("3 - Multiplicar")
    print("4 - Dividir")
    print("5 - Porcentagem")
    print("6 - Ver historico")
    print("7 - Sair")

    # O programa pausa e espera o usuário digitar alguma coisa.
    # O número que o usuário digitar vai ser guardado na variável 'opcao'.
    opcao = input("Sua escolha: ")


    # DECISÃO (IF e ELIF)

    # saida opcao 7:
    if opcao == "7":
        print("Tchau!")
        # a chave para "nao". quando ler tudo, desliga o programa.
        programa_rodando = "nao"

    # historico opcao 6:
    elif opcao == "6":
        print(" ")
        print("--- HISTORICO ---")
        print("Espaco 1:", mem_1)
        print("Espaco 2:", mem_2)
        print("Espaco 3:", mem_3)
        print("Espaco 4:", mem_4)
        print("Espaco 5:", mem_5)
        print("---------------------")

    # agrupar opções.
    # Se escolher qualquer número de 1 a 5, o programa entra neste bloco
    elif opcao == "1" or opcao == "2" or opcao == "3" or opcao == "4" or opcao == "5":

        # dois números pra conta
        # O 'float' serve pra avisar que pode ter numeros quebrados"quebrado" (ter decimais).
        n1 = float(input("Primeiro numero: "))
        n2 = float(input("Segundo numero: "))

        # Cria uma caixa zerada apenas para existir antes de descobrirmos o resultado verdadeiro.
        resultado = 0

        # escolhe qual das opcoes foram escolhidas e aplica a matematica
        if opcao == "1":
            resultado = n1 + n2
        elif opcao == "2":
            resultado = n1 - n2
        elif opcao == "3":
            resultado = n1 * n2
        elif opcao == "4":
            resultado = n1 / n2
        elif opcao == "5":
            resultado = (n1 * n2) / 100

        # mostra a resposta da conta feita abaixo
        print("Resultado:", resultado)


        # sistema de historico automatico
        # O programa olha para a variável 'lugar_da_memoria' para saber em qual gaveta guardar.

        # apontando para a posição 1
        if lugar_da_memoria == 1:
            mem_1 = resultado        # Ele guarda o resultado novo na gaveta 1.
            lugar_da_memoria = 2     # manda para o 5 (preparando-se para a próxima conta).

        # estará no 2
        elif lugar_da_memoria == 2:
            mem_2 = resultado        # Então ele guarda na gaveta 2.
            lugar_da_memoria = 3    # manda para o 3

        # Na conta seguinte, estará no 3
        elif lugar_da_memoria == 3:
            mem_3 = resultado        # Guarda na gaveta 3.
            lugar_da_memoria = 4     # manda para o 4

        # Na conta seguinte, a seta estará no 4...
        elif lugar_da_memoria == 4:
            mem_4 = resultado        # Guarda na gaveta 4.
            lugar_da_memoria = 5     # manda para o 5

        # Quando a seta chega no 5, significa que vamos encher a última gaveta
        elif lugar_da_memoria == 5:
            mem_5 = resultado
            lugar_da_memoria = 1     # volta a apontar para o 1
                                     # Com isso, a 6 conta da calculadora vai voltar la no 'if lugar_da_memoria == 1'
                                     # e vai APAGAR a conta que estava na 'mem_1', recomeçando o ciclo
```
